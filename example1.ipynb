```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
import numpy as np
```

```python
X, _ = make_moons(n_samples=1000)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
results = kmeans.fit_predict(X)
classesValues =kmeans.labels_
```

```python
c1,c2=[],[]
for index,classes in enumerate(classesValues):
    if classes==0:
        c1.extend(X[index])
    else :
        c2.extend(X[index])
```

```python
classter1=np.array(c1).reshape(499,2)
classter2=np.array(c2).reshape(501,2)

plt.scatter(classter1[:,0],classter1[:,1])
plt.scatter(classter2[:,0],classter2[:,1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan1 = DBSCAN(eps=0.3, min_samples=3)
dbscan1.fit(classter1)
plt.scatter(classter1[:,0],classter1[:,1], c=dbscan1.labels_)

plt.show()
```

```python
dbscan2 = DBSCAN(eps=0.2, min_samples=5)
dbscan2.fit(classter2)
plt.scatter(classter2[:,0],classter2[:,1], c=dbscan2.labels_)

plt.show()
```

```python
plt.scatter(classter1[:,0],classter1[:,1], c=dbscan1.labels_)

plt.scatter(classter2[:,0],classter2[:,1], c=dbscan2.labels_)

plt.show()
```
